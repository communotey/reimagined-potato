```python
with open('course_list.txt', 'r') as f:
    x = f.readlines()
courses = []
for items in x:
    items = items.rstrip('\n')
    #items = json.loads(items)
    courses.append(items)
```

```python
json_course_list = []

for json_item in courses:
    course_schema = {
                    "department":"Math",
                    "code":"math-1za3",
                    "title":"MATH 1ZA3",
                    "semesterId":"1",
                    "courseLevel":"1",
                    "group":"",
                    "_school":ObjectId("5b0b3460c83c81aa41a16e92"),
                    "_creator":ObjectId("5b0b306618d0145954a2c011"),
                    "createdAt":datetime.utcnow(),
                    "name":""
                    }
    
    
    m = re.search(r"\d",json_item)
    m = m.start()
    course_schema['department'] = json_item[:m].strip().upper()
    
    course_schema['code'] = str(json_item.replace(" ","-").lower()).replace("&","and")
    course_schema['title'] = json_item
    course_schema['course_level'] = json_item[-4]
    course_schema['name'] = json_item
    json_course_list.append(course_schema)
    
```
